```
import os
import pandas as pd
import glob


print(os.listdir("DATA"))
```

Merging csv files

```
# Path to all CSV files in data folder
files = glob.glob("DATA/*.csv")

dfs = []
for file in files:
    df = pd.read_csv(file)
    dfs.append(df)

# Merge all CSVs into one DataFrame
merged_df = pd.concat(dfs, ignore_index=True)

# Display results
print("Merged shape:", merged_df.shape)
merged_df.head(15)
```
